## 4. 删除'Courier Status'列中取值为Cancelled和NaN的订单（0.5分）

```python
#在del_储存Courier Status'列中取值为Cancelled和NaN的订单
del_=sale_data[sale_data['Courier Status'] == 'Cancelled'].index # 去除Cancelled值
sale_data.drop(del_,inplace=True) #去除符合del_条件的数据 
sale_data=sale_data.dropna(subset=['Courier Status'])#去除Nan值
sale_data
```

## 9. B2B和非B2B用户的购买力如何？


### a. 统计两类用户的总销售金额和总销量。（1分）

```python
sale_data['B2B'][sale_data['B2B']==True] = 'B2B'
sale_data['B2B'][sale_data['B2B']==False] = 'Non-B2B'
```

```python
sale_data.groupby('B2B')[['Amount','Qty']].sum()
```

### b. 分别统计两类用户畅销TOP5（销售金额、销量）的商品有哪些。（1.5分）

```python
#对B2B和SKU进行Grouping,然后求产品的销量，由于sum()后按Qty的标准来排序，因此再次进行grouping，然后nlargest(5)来表示前5个项目
sale_data.groupby(['B2B','SKU'])['Qty'].sum().groupby(level=0, group_keys=False).nlargest(5).to_frame()
```

```python
#与上所述的方法一样进行求各项目的和
sale_data.groupby(['B2B','SKU'])['Amount'].sum().groupby(level=0, group_keys=False).nlargest(5).to_frame()
```

## 分别求

## B2B类

### 销售金额

```python
#先求相同b2b类的各商品的销售金额总和
b=sale_data.groupby(['B2B','SKU'])[['Amount']].sum()
#为了排序index变成column
b.reset_index(inplace=True)
#b2b类按销售金额进行排序
b[b['B2B']=='B2B'].sort_values(['Amount'],ascending =False).set_index('B2B')
```

### 销量

```python
#与上所述的方法一样求各商品销售量的排名
b=sale_data.groupby(['B2B','SKU'])[['Qty']].sum()
b.reset_index(inplace=True)
b[b['B2B']=='B2B'].sort_values(['Qty'],ascending =False).set_index('B2B')
```

## Non-B2B类

```python
### 销售金额
```

```python
b=sale_data.groupby(['B2B','SKU'])[['Amount']].sum()
b.reset_index(inplace=True)
b[b['B2B']=='Non-B2B'].sort_values(['Amount'],ascending =False).set_index('B2B')
```

### 销量

```python
b=sale_data.groupby(['B2B','SKU'])[['Qty']].sum()
b.reset_index(inplace=True)
b[b['B2B']=='Non-B2B'].sort_values(['Qty'],ascending =False).set_index('B2B')
```

## 10.分别统计4-6月订单销量TOP3的商品以及各月销售金额TOP3的品类（1.5分）

```python
#先为了按月份分类形成相关的函数
trans=lambda x : x[:2]
#利用map在Date column 应用这一函数取出月份
sale_data['Date']= sale_data['Date'].map(trans)
#利用groupby,grouping Date 与 SKU cloumn 然后求Qty的总和,然后直接排序的话，不管月份从大到小排序，因此再次进行grouping 
#LEVEL指想grouping的index的序号，使用group_keys=False防止出现重复的情况，最后用Nlargest输出前3的品类
sale_data.groupby(['Date','SKU'])['Qty'].sum().groupby(level=0, group_keys=False).nlargest(3).to_frame()





```
